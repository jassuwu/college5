```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xy = [(4,33), (4.5, 42), (5, 45), (5.5,51), (6, 53), (6.5, 61), (7, 62)]
```

```python
X = np.array([[1,i] for i in [i[0] for i in xy]])
Y = ([ i for i in [i[1] for i in xy]])
print(X, Y)
```

```python
learning_rate = 0.0088315
iterations = 0
theta = [0,0]
thetaStar = [0,0]
while True:
    grad = np.zeros((X.shape[1],1))

    for xi, yi in xy:
        grad[0] += yi - theta[0] - theta[1]*xi
        grad[1] += (yi - theta[0] - theta[1]*xi)*xi

    thetaStar[0] = theta[0] + learning_rate*grad[0]
    thetaStar[1] = theta[1] + learning_rate*grad[1]

    iterations+=1
    if thetaStar == theta:
        break

    theta[0] = thetaStar[0]
    theta[1] = thetaStar[1]

print(theta)
```

```python
estimatorFn = theta[1]*[xi for xi,yi in xy ] + theta[0]
plt.plot([xi for xi,yi in xy ], estimatorFn, color='red')
plt.scatter([xi for xi,yi in xy ],Y)
plt.show()
```
