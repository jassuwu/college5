```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
xy = [(4,33), (4.5, 42), (5, 45), (5.5,51), (6, 53), (6.5, 61), (7, 62)]
X = []
Y = []
for x,y in xy:
    X.append(x)
    Y.append(y)
X = np.array(X)
Y = np.array(Y)
```

```python
eta = 0.0088315
Bold = np.array([0,0])
Bnew = np.array([0,0])
iterations = 0
plt.ylim((30, 70))
plt.plot(X, Y, 'bo')
estimatorFn = Bnew[0] + Bnew[1]*X
```

```python
while True:
    grad = np.array([0,0])
    for xi, yi in xy:
        grad[0] += (yi - Bold[0] - Bold[1]*xi)
        grad[1] += -xi*(yi - Bold[0] - Bold[1]*xi)
    
    Bnew[0] = Bold[0] - eta*grad[0]
    Bnew[1] = Bold[1] - eta*grad[1]
    iterations += 1
    plt.plot(X, estimatorFn, 'r-')
    plt.show()
    if np.array_equal(Bold, Bnew):
        break

    Bold = Bnew

print(Bnew)
print(iterations)

# estimatorFn = Bnew[0] + Bnew[1]*X
# plt.ylim((30, 70))
# plt.plot(X, estimatorFn, 'r-')
# plt.plot(X, Y, 'bo')
# plt.show()
```
