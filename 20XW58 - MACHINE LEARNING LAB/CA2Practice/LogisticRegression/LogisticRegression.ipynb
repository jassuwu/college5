```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, normalize
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import chi2, SelectKBest
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('wbc.csv')

df = df.drop(['id', 'Unnamed: 32'], axis=1)

X = df.drop(['diagnosis'], axis=1)
y = df['diagnosis']

y = LabelEncoder().fit_transform(y)

print(X)
X = pd.DataFrame(normalize(X), columns=X.columns)
print(X)

bestFeatures = SelectKBest(score_func=chi2, k=10)
fit = bestFeatures.fit(X,y)
bfScores = pd.DataFrame(fit.scores_)
bfColumns = pd.DataFrame(X.columns)
featureScores = pd.concat([ bfColumns, bfScores], axis=1)
featureScores.columns = ['Specs', 'Score']
print(featureScores.nlargest(10, 'Score'))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("AccScore", accuracy_score(y_test, y_pred))
print("ConMat", confusion_matrix(y_test, y_pred))
print("ClassRep", classification_report(y_test, y_pred))
```
