```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score

bc = pd.read_csv('wbc_original.csv')

bc = bc.drop(['id', 'bare_nucleoli'], axis=1)
X = bc.drop(['class'], axis=1)
y = bc['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

DTC = DecisionTreeClassifier()
DTC.fit(X_train, y_train)
y_pred = DTC.predict(X_test)

print("AccSco: ", accuracy_score(y_test, y_pred))
```

```python
pd.DataFrame(DTC.feature_importances_, index=X.columns).plot(kind='bar')
```

```python

fig = plt.figure(figsize=(25,20))
_ = plot_tree(DTC, feature_names=X.columns, class_names={0:"Malignant", 1:"Benign"}, filled=True, fontsize=12)
```
